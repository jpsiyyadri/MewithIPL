```python
import sys
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.patches as patches
from ipywidgets import interact, widgets
```

```python
ipl_matches = pd.read_csv("./ipl/ipl_matches.csv")
ipl_deliveries = pd.read_csv("./ipl/ipl_deliveries.csv")
```

```python
%matplotlib inline
def get_stat(df, team, season):
    #     Team data
    team_ipl = df[(df['team1']==team) | (df['team2']==team)]
    #     Season data
    season_ipl = team_ipl[team_ipl['season'] == season]
    #     all wins
    team_won = season_ipl[season_ipl['winner'] == team]
    #     toss won
    team_toss = season_ipl[season_ipl['toss_winner'] == team]
    #     toss won and match won
    team_toss_won = team_toss[team_toss['winner'] == team]
    #     toss won and choose to bat
    team_toss_bat = team_toss[team_toss['toss_decision'] == 'bat']
    #     toss won and choose to bat and match won
    team_toss_bat_won = team_toss_bat[team_toss_bat['winner'] == team]

    y = ['matches', 'won', 'toss', 'toss & match', 'toss & bat', 'toss & bat & match' ]
    x = [len(season_ipl), len(team_won), len(team_toss), len(team_toss_won), len(team_toss_bat), len(team_toss_bat_won)]

    y_pos = np.arange(len(y))
#     print(y_pos)
    y_pos = [0,0.6,1.2,1.8,2.4,3.0]
    plt.clf()
    plt.barh(y_pos, x, align='center', alpha=0.5, height=0.3)
    plt.xticks(np.arange(0, 21, step=1))
    plt.yticks(y_pos, y)
    plt.xlabel('performance')
    plt.title('Season '+str(season)+'\ndecision v/s wins')
    plt.grid(True)
    plt.figure()
    
```

```python
from ipywidgets import *
import numpy as np
df = ipl_matches

#identifies the columns in the dataframe
seasons = list(df.season.unique())
teams = list(df[df['season']==seasons[0]].team1.unique())

#plot function
def set_plot(seasons_dd, teams_dd):
    data = df
    x = int(seasons_dd)
    get_stat(ipl_matches, teams_dd, x)

#function to specify the type of plot based on users input
def set_plotType(_):
    data = df
    x = int(seasons_dd.value)
    teams = list(data[data['season']==x].team1.unique())
    teams_dd.options = list(teams)


#define widgets
seasons_dd = widgets.Dropdown(
    options=seasons,
    value=seasons[0],
    description='Season'
)

seasons_dd.set_title  = 'seasons_dd'

teams_dd = widgets.Dropdown(
    options=teams,
    value=teams[0],
    description='Team'
)

teams_dd.set_title  = 'teams_dd'

# change in drop down
seasons_dd.observe(set_plotType)
teams_dd.observe(set_plotType)


#interact function
interact(set_plot, seasons_dd = seasons_dd, teams_dd = teams_dd)
```
